# Processing Data

```python
diabetes.dtypes
```

```python
diabetes = diabetes.drop(columns=['cholesterol'])
diabetes
#I drop the cholesterol to improve the model
```

```python
X = diabetes.drop(['diabetes'],axis=1)
y = diabetes['diabetes']
diabetes
```

```python
#we only have numerical columns so we dont need to separate into x and y (X_train_cat , X_train_num , X_test_cat , X_test_num)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
#X_train = X_train.select_dtypes(include = np.number)
#X_test = X_test.select_dtypes(include = np.number)
```

```python
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_train)
X_train_scaled = transformer.transform(X_train)
print(X_train_scaled.shape)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_train_scaled
```

```python
X_test_scaled = transformer.transform(X_test)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
X_test_scaled
```

# Imbalance and metrics

```python
count_classes = diabetes['diabetes'].value_counts()
count_classes
count_classes.plot(kind = 'bar')
```

# Logistic Regression Model

```python
#Evaluate the model.
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_scaled, y_train)
LR.score(X_test_scaled, y_test) 
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

pred = LR.predict(X_test_scaled)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

# Hypotesis Testing

```python
from scipy.stats import ttest_1samp
```

```python
age_values = diabetes['age'].dropna()
```

```python
mean = 40
age_values = diabetes.loc[diabetes['diabetes'] == 1, 'age'].dropna()
t_stat, p_value = ttest_1samp(age_values, mean)
alpha = 0.05

print('stat is  ', t_stat)
print('pvalue for the two-tailed test is ', p_value)


if p_value < alpha:
    print("Reject the null hypothesis: The mean age of individuals with diabetes is significantly different from 40")
else:
    print("Fail to reject the null hypothesis: The mean age of individuals with diabetes is not significantly different from 40 years")
```
